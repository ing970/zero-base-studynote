# 40_[연습문제] 함수(01)

- 다음과 같이 출력 될 수 있도록 산술연산 계산기를 함수를 이용해서 만들어 보자.

```python
def cal():
    
    def add(n1, n2):
        print(f"{n1} + {n2} = {n1 + n2}")
        
    def sub(n1, n2):
        print(f"{n1} - {n2} = {n1 - n2}")
    
    def mul(n1, n2):
        print(f"{n1} * {n2} = {n1 * n2}")
    
    def div(n1, n2):
        print(f"{n1} / {n2} = {n1 / n2}")
    
    def rest(n1, n2):
        print(f"{n1} % {n2} = {n1 % n2}")
    
    def share(n1, n2):
        print(f"{n1} // {n2} = {n1 // n2}")
    
    def sqare(n1, n2):
        print(f"{n1} ** {n2} = {n1 ** n2}")
        
        
    cal = int(input("1.덧셈  2.뺄셈  3.곱셈  4.나눗셈  5.나머지  6.몫  7.제곱승  8.종료  "))
    n1 = float(input("첫 번째 숫자 입력: "))
    n2 = float(input("두 번째 숫자 입력: "))
    
    if cal == 1:
        add(n1, n2)
    elif cal == 2:
        sub(n1, n2)
    elif cal == 3:
        mul(n1, n2)
    elif cal == 4:
        div(n1, n2)
    elif cal == 5:
        rest(n1, n2)
    elif cal == 6:
        share(n1, n2)
    elif cal == 7:
        sqare(n1, n2)
    elif cal == 8:
        print("Bye~~")

cal()
        
```

# 41_[연습문제] 함수(02)

- 다음과 같이 출력 될 수 있도록 이동거리와 이동시간을 반환하는 함수를 만들어 보자.

```python
# 다음과 같이 출력 될 수 있도록 이동거리와 이동시간을 반환하는 함수를 만들어 보자.
def get_distance(s, h, m):
    
    distance = s * (h + (m / 60))
    
    return distance

s = float(input("속도(km/h) 입력: "))
h = float(input("시간(h) 입력: "))
m = float(input("시간(m) 입력: "))
d = get_distance(s, h, m)

print(f"{s}(km/h)속도로 {h}(h)시간 {m}(m)분 동안 이동한 거리: {d}(km)")
```

```python
# 다음과 같이 출력 될 수 있도록 이동거리와 이동시간을 반환하는 함수를 만들어 보자.
def get_time(s, d):
    
    result = d / s
    h = int(result)
    m = int((result - h) * 60)
    l = [h, m]
    
    return l
    
s = float(input("속도(km/h) 입력: "))
d = float(input("거리(km) 입력: "))
t = get_time(s, d)
print(f"{s}(km/h)속도로 {d}(km) 이동한 시간: {t[0]}(h)시간 {t[1]}(m)분")
```

# 42_[연습문제] 함수(03)

```python
# 다음과 같이 출력 될 수 있도록 비행기 티켓 영수증 출력 함수를 만들어 보자.

def bill():
    
    def discount(price):
        result = price * 0.5 
        return int(result)
    
    child_price = 18000
    infant_price = 25000
    adult_price = 50000
 
    child = int(input("유아 입력: "))
    dis_child = int(input("할인대상 유아 입력: "))
    infant = int(input("소아 입력: "))
    dis_infant = int(input("할인대상 소아 입력: "))
    adult = int(input("성인 입력: "))
    dis_adult = int(input("할인대상 성인 입력:"))
    
    ch_fee = (child * child_price)
    dis_ch_fee = dis_child * discount(child_price)
    in_fee = infant * infant_price
    dis_in_fee = dis_infant * discount(infant_price)
    adult_fee = adult * adult_price
    dis_adult_fee = dis_adult * discount(adult_price)
    total_fee = ch_fee + dis_ch_fee + in_fee + dis_in_fee + adult_fee + dis_adult_fee
    
    print("="*40)
    print(f"유아 {child}명 요금: {ch_fee:,}원")
    print(f"유아 할인 대상 {dis_child}명 요금: {dis_ch_fee:,}원")
    print(f"소아 {infant}명 요금: {in_fee:,}원")
    print(f"소아 할인 대상 {dis_infant}명 요금: {dis_in_fee:,}원")
    print(f"성인 {adult}명 요금: {adult_fee:,}원")
    print(f"성인 할인 대상 {dis_adult}명 요금: {dis_adult_fee:,}원")
    print("="*40)
    print(f"Total: {child + dis_child + infant + dis_infant + adult + dis_adult}명")
    print(f"Total_price: {total_fee:,}원")
    print("="*40)
    
    
bill()
```

# 43_[연습문제] 함수(04)

```python
# 다음과 같이 출력 될 수 있도록 재귀함수를 이용해서 팩토리얼 함수를 만들어보자.
def factorial(n):
    if n <= 1:
        return 1
    else:
        return n * factorial(n - 1)
        
factorial(10)
```

# 44_[연습문제] 함수(05)

```python
# 다음과 같이 출력 될 수 있도록 등차 수열의 n번째 값과 합을 출력하는 함수를 만들어보자
def cal(n1, d, n):
    value_n = 0; sum_n = 0;
    
    i = 1
    while i <= n:
        
        if i == 1:
            value_n = n1
            sum_n += value_n
            print(f"{i}번째 항의 값: {value_n}")
            print(f"{i}번째 항까지의의 핪: {sum_n}")
            
            i += i
            continue
            
        value_n += d
        sum_n += value_n
        print(f"{i}번째 항의 값: {value_n}")
        print(f"{i}번째 항까지의의 핪: {sum_n}")
        
        i += 1

input_n1 = int(input("a1 입력: "))
input_d = int(input("공차 입력: "))
input_n = int(input("n 입력: "))

cal(input_n1, input_d, input_n)
```

# 47_[연습문제] 모듈(02)

- 상품 구매 개수에 따라 할인율이 결정되는 모듈을 만들고, 다음과 같이 계산 결과가 출력되는 프로그램을 만들어보자.



```python
choice_cnt = 0
sum_price = 0

while True:
    choice = int(input("상품을 구매 하시겠어요? 1.구매  2.종료"))
    
    if choice == 1:
        price = int(input("상품 가격 입력: "))
        sum_price += price
        choice_cnt += 1
    elif choice == 2:
        break

discount = min(choice_cnt * 5, 25)
total = sum_price - int(sum_price * (discount / 100))
 
print(f"할인율: {discount}%")
print(f"합계: {total:,}원")
```

# 48_[연습문제] 모듈(03)

## - 로또 모듈을 만들고 다음과 같이 로또 결과가 출력될 수 있도록 프로그램을 만들어보자.




```python
import random

# 사용자로부터 6개의 번호를 입력받음
chosen_numbers = [int(input(f"번호(1 ~ 45) 입력: ")) for _ in range(6)]

# 로또 번호 6개와 보너스 번호 1개를 무작위로 생성
lotto_numbers = random.sample(range(1, 46), 7)

# 보너스 번호를 설정하고 로또 번호에서 제거
bonus_number = lotto_numbers.pop()

# 선택한 번호와 로또 번호를 비교하여 일치하는 번호를 찾음
matching_numbers = [num for num in chosen_numbers if num in lotto_numbers]

print("아쉽습니다. 다음 기회에~")
print(f"기계 번호: {sorted(lotto_numbers)}")
print(f"보너스 번호: {bonus_number}")
print(f"선택 번호: {sorted(chosen_numbers)}")
print(f"일치 번호: {sorted(matching_numbers)}")
```

# 49_[연습문제] 모듈(04)

- 순열 계산 모듈을 만들고 다음 순열 계산 결과를 출력해 보자.

```python
import itertools
import math

num_n = int(input("num_n 입력: "))
num_r = int(input("num_r 입력: "))

print("n :", num_n)
while num_r > 1:
    num_n -= 1
    num_r -= 1
    print("n :", num_n)

num_n = int(input("num_n 입력: "))
num_r = int(input("num_r 입력: "))

permutations = list(itertools.permutations(range(1, num_n+1), num_r))

print(f"{num_n}P{num_r}: {len(permutations)}")
print(f"{num_n}P{num_r} 개수: {math.factorial(num_n) // math.factorial(num_n - num_r)}")
print(*permutations, sep=", ")
```
