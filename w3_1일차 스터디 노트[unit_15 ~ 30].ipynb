# 21_·Ñë·Öµ·Ñá·Ö©·ÑÇ·Ö°·Ñé·Öµ ·Ñâ·ÖÆ·Ñã·Öß·ÜØ

- ÏÑ∏ Î≤àÏß∏ Ìï≠ÏùÄ Îëê Î≤àÏß∏ Ìï≠Í≥º Ï≤´ Î≤àÏß∏ Ìï≠ÏùÑ ÎçîÌïú Ìï©Ïù¥Îã§.

```python
def fibonacci(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        a, b = 0, 1
        for _ in range(n-1):
            a, b = b, a + b
        return b
    
    
for i in range(1, 10):
    print(fibonacci(i))
```

# 30_·Ñí·Ö™·Ü®·ÑÖ·Ö≤·ÜØ(·Ñë·Ö°·Ñã·Öµ·Ñä·Ö•·Ü´)

- Î∞ïÏä§Ïóê ‚ÄòÍΩù‚ÄôÏù¥ Ï†ÅÌûå Ï¢ÖÏù¥Í∞Ä 4Ïû• ÏûàÍ≥†, ‚ÄòÏÑ†Î¨º‚ÄôÏù¥ Ï†ÅÌûå Ï¢ÖÏù¥Í∞Ä 3Ïû•Ïù¥ ÏûàÏùÑ Îïå,
- ÌååÏù¥Ïç¨ÏùÑ Ïù¥Ïö©Ìï¥ÏÑú ‚ÄòÍΩù‚Äô2Ïû•Í≥º ‚ÄòÏÑ†Î¨º‚Äô 1Ïû•ÏùÑ ÎΩëÎäî ÌôïÎ•†(%)ÏùÑ Ï∂úÎ†•ÌïòÏûê.

```python
def pro_fun():
    num_n = int(input("num_n ÏûÖÎ†•: "))
    num_r = int(input("num_r ÏûÖÎ†•: "))


    result_p = 1 # ÏàúÏó¥
    result_r = 1
    result_c = 1 # Ï°∞Ìï©


    for i in range(num_n, num_n - num_r, -1):
        result_p *= i
    print(f"{result_p = }")

    for i in range(num_r, 0, -1):
        result_r *= i
    print(f"{result_r = }")

    result_c = int(result_p / result_r)
    print(f"{result_c = }")
    
    return result_c

sample = pro_fun()
print(f"{sample = }")
print("=" * 30)
event1 = pro_fun()
print(f"{event1 = }")
print("=" * 30)
event2 = pro_fun()
print(f"{event2 = }")
print("=" * 30)
probability = (event1 * event2) / sample
print(f"probability: {round(probability * 100, 2)}")
```
