````python
# 2-1
# í•¨ìˆ˜ ì •ì˜
def get_region(gu):
    """
    guë¥¼ ë°›ì•„ region_dictì˜ valuesì¸ "êµ¬"ë‘ ì¼ì¹˜í•˜ë©´,
    í•´ë‹¹í•˜ëŠ” dictì˜ keyì¸ regionì„ ë°˜í™˜í•©ë‹ˆë‹¤.
    """
    for region, guu in region_dict.items():
        if gu in guu:
            return region
    return None
````
