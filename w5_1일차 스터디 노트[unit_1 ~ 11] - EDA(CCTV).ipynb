### üìç df.sort_values( )
- Pandas DataFrameÏùò sort_values() Î©îÏÜåÎìúÎäî DataFrameÏùÑ ÌïòÎÇò Ïù¥ÏÉÅÏùò Ïó¥ÏùÑ Í∏∞Ï§ÄÏúºÎ°ú Ï†ïÎ†¨ÌïòÎäî Îç∞ ÏÇ¨Ïö©Îê©ÎãàÎã§. 
- Ïù¥ Î©îÏÜåÎìúÎäî Ï†ïÎ†¨Îêú DataFrameÏùÑ Î∞òÌôòÌïòÎ©∞ ÏõêÎûò DataFrameÏùÄ Î≥ÄÍ≤ΩÎêòÏßÄ ÏïäÏäµÎãàÎã§.
~~~python
df.sort_values(by, axis=0, ascending=True, inplace=False, kind='quicksort', na_position='last')
~~~
>- by: ÌïòÎÇòÏùò Ïó¥ ÎòêÎäî Ïó¥ Î¶¨Ïä§Ìä∏Î•º ÎÇòÌÉÄÎÇ¥Î©∞ Ïù¥Î•º Í∏∞Ï§ÄÏúºÎ°ú DataFrameÏù¥ Ï†ïÎ†¨Îê©ÎãàÎã§.
>- axis: Í∏∞Î≥∏Í∞íÏùÄ 0ÏûÖÎãàÎã§. Ïó¥ÏùÑ Í∏∞Ï§ÄÏúºÎ°ú Ï†ïÎ†¨Ìï©ÎãàÎã§. 1Î°ú ÏÑ§Ï†ïÌïòÎ©¥ ÌñâÏùÑ Í∏∞Ï§ÄÏúºÎ°ú Ï†ïÎ†¨Ìï©ÎãàÎã§.
>- ascending: Í∏∞Î≥∏Í∞íÏùÄ TrueÏûÖÎãàÎã§. TrueÎ°ú ÏÑ§Ï†ïÌïòÎ©¥ Ïò§Î¶ÑÏ∞®ÏàúÏúºÎ°ú Ï†ïÎ†¨ÌïòÍ≥†, FalseÎ°ú ÏÑ§Ï†ïÌïòÎ©¥ ÎÇ¥Î¶ºÏ∞®ÏàúÏúºÎ°ú Ï†ïÎ†¨Ìï©ÎãàÎã§.
>- inplace: Í∏∞Î≥∏Í∞íÏùÄ FalseÏûÖÎãàÎã§. TrueÎ°ú ÏÑ§Ï†ïÌïòÎ©¥ ÏõêÎ≥∏ DataFrameÏù¥ Î≥ÄÍ≤ΩÎêòÍ≥† Î∞òÌôò Í∞íÏù¥ ÏóÜÏäµÎãàÎã§. FalseÎ°ú ÏÑ§Ï†ïÌïòÎ©¥ ÏõêÎ≥∏ DataFrameÏùÄ Í∑∏ÎåÄÎ°ú ÎëêÍ≥† Ï†ïÎ†¨Îêú DataFrameÏù¥ Î∞òÌôòÎê©ÎãàÎã§.
>- kind: Í∏∞Î≥∏Í∞íÏùÄ 'quicksort'ÏûÖÎãàÎã§. Ï†ïÎ†¨ ÏïåÍ≥†Î¶¨Ï¶òÏùÑ ÏÑ†ÌÉùÌï©ÎãàÎã§. 'quicksort', 'mergesort', 'heapsort' Ï§ë ÏÑ†ÌÉùÌï† Ïàò ÏûàÏäµÎãàÎã§.
>- na_position: Í∏∞Î≥∏Í∞íÏùÄ 'last'ÏûÖÎãàÎã§. Í≤∞Ï∏°Í∞í(NaN)Ïùò ÏúÑÏπòÎ•º ÏÑ†ÌÉùÌï©ÎãàÎã§. 'last'Îäî Í≤∞Ï∏°Í∞íÏùÑ Îß® ÎßàÏßÄÎßâÏóê ÎëêÍ≥†, 'first'Îäî Í≤∞Ï∏°Í∞íÏùÑ Îß® ÏïûÏóê Îë°ÎãàÎã§.
