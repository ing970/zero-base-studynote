# 38. Î∞òÎ≥µÎ¨∏ Ï†úÏñ¥( continue )

- Î∞òÎ≥µ Ïã§Ìñâ Ï§ë continueÎ•º ÎßåÎÇòÎ©¥ Ïã§ÌñâÏùÑ ÏÉùÎûµÌïòÍ≥†, Îã§Ïùå Î∞òÎ≥µ Ïã§ÌñâÎ¨∏ÏúºÎ°ú ÎÑòÏñ¥Í∞ÑÎã§.

```python
for i in range(1, 50):
    if i % 7 != 0:
        continue
        
    print(f"{i}Îäî 7Ïùò Î∞∞ÏàòÏûÖÎãàÎã§.")
```

```python
# elseÏùò Ïã§ÌñâÎ¨∏ÏùÄ Î∞òÎ≥µÎ¨∏Ïù¥ Ï¢ÖÎ£åÎêú ÌõÑ Ïã§ÌñâÎêúÎã§. Í∂ÅÍ∏à!!

cnt = 0
for i in range(1, 100):
    if i % 7 != 0:
        continue
    
    print(f"{i}Îäî 7Ïùò Î∞∞ÏàòÏù¥Îã§.")
    cnt += 1

else:
    print(f"99ÍπåÏßÄÏùò Ï†ïÏàò Ï§ë 7Ïùò Î∞∞ÏàòÎäî {cnt}Í∞ú Ïù¥Îã§.")
```

- 1Î∂ÄÌÑ∞ 100ÍπåÏßÄÏùò Ï†ïÏàò Ï§ë 3Í≥º 7Ïùò Í≥µÎ∞∞ÏàòÏôÄ ÏµúÏÜå Í≥µÎ∞∞ÏàòÎ•º Ï∂úÎ†•ÌïòÏûê.

```python
min_num = 0

for i in range(1, 100):
    if i % 3 != 0 or i % 7 != 0:
        continue
    
    # 3Í≥º 7Ïùò Í≥µÎ∞∞Ïàò Ï∂úÎ†•
    print(f"Í≥µÎ∞∞Ïàò: {i}")
    
    # 3Í≥º 7Ïùò Í≥µÎ∞∞Ïàò Ï§ë ÏµúÏ¥àÏùò Í≥µÎ∞∞ÏàòÎßå Î∞òÎ≥µÎ¨∏ Î∞ñÏùò min_numÏóê Ï†ïÏùòÎêúÎã§. 
    if min_num == 0:
        min_num = i

else:
    print(f"ÏµúÏÜå Í≥µÎ∞∞Ïàò: {min_num}")
```
