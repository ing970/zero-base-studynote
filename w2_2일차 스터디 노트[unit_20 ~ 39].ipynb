# 20_객체와 메모리

- 변수는 객체의 메모리 주소를 저장하고 이를 이용해서 객체를 참조한다.

### 📎 실습
- 국어, 영어, 수학 점수를 입력받아 리스트에 저장하고 원본을 유지한 상태로,
- 복사본을 만들어 과목별 점수를 10% 올렸을 경우에 평균을 출력해 보자.

````python
kor = int(input("국어 점수 입력: "))
eng = int(input("영어 점수 입력: "))
mat = int(input("수학 점수 입력: "))

score_l = [kor, eng, mat]
copy_l = score_l.copy()

for idx, score in enumerate(copy_l):
    # 10% 상승 금액
    result = score * 1.1
    # 상승된 금액이 100을 넘어가면 100으로 설정. and copy_l에 넣기
    copy_l[idx] = 100 if result > 100 else result
    
print(f"이전 평균: {round(sum(score_l) / len(score_l), 2)}")
print(f"이후 평균: {round(sum(copy_l) / len(copy_l), 2)}")
````

# 21_얕은복사와 깊은복사

### 📍얕은 복사
- 얕은 복사란, 객체 주소를 복사하는 것으로 객체 자체가 복사되지 않는다.
### 📍깊은 복사
- 깊은 복사란, 객체 자체를 복사하는 것으로 또 하나의 객체가 만들어진다.

### 📎 실습
- 선수의 원본 점수를 이용해서 평균을 출력하고
- 최고값과 최저값을 제외한 평균을 출력하는 프로그램을 만들어보자.

````python
l = [7.9, 8.3, 8.7, 8.8, 8.9, 9.0, 9.1, 9.5]
copy_l = l.copy()

copy_l.sort()
copy_l.pop(0)
copy_l.pop()

o_total = round(sum(l), 2)
o_avg = round(o_total / len(l), 2)
print(f"original total: {o_total}")
print(f"original average: {o_avg}")

c_total = round(sum(copy_l), 2)
c_avg = round(c_total / len(copy_l), 2)
print(f"copy total: {c_total}")
print(f"copy avg: {c_avg}")

print(f"ori_avg - copy_avg: {o_avg - c_avg}")
````

# 22_클래스 상속

### 📍상속 이란
- 클래스는 또 다른 클래스를 상속해서 내 것처럼 사용할 수 있다.

### 📎 실습
- 덧셈, 뺄셈 기능이 있는 클래스를 만들고
- 이를 상속하는 클래스를 만들어서 곱셈과 나눔셈 기능을 추가해 보자.

````python
class Calculator_parent:
    
    def add(self, n1, n2):
        return n1 + n2
    
    def sub(self, n1, n2):
        return n1 - n2
````

````python
# child 클래스는 parent 클래스를 상속한다.
class Calculator_child(Calculator_parent):
    
    def mul(self, n1, n2):
        return n1 * n2
    
    def div(self, n1, n2):
        return n1 / n2
````

````python
cal = Calculator_child()

print(f"cal add: {cal.add(4, 6)}")
print(f"cal sub: {cal.sub(4, 6)}")
print(f"cal mul: {cal.mul(4, 6)}")
print(f"cal div: {cal.div(4, 6)}")
````

### 📍 \_\_init\_\_(  )
- \_\_init\_\_( )가 속성을 초기화 한다.

````python
class Child:
    
    def __init__(self, cn1, cn2):
        print("print child numbers")
        self.cn1 = cn1
        self.cn2 = cn2

ch = Child(40, 50)
print(f"ch.cn1 : {ch.cn1}")
print(f"ch.cn2 : {ch.cn2}")
````

### 📍 super(  )
- 상위 클래스의 속성을 초기화하기 위해서 super( )를 이용한다.

````python
class Child:
    
    def __init__(self, cn1, cn2):
        print("print child numbers")
        self.cn1 = cn1
        self.cn2 = cn2


class Parent(Child):
    
    def __init__(self, pn1, pn2):
        print("print parent numbers")
        
        # Child 클래스의 __init__ 메서드를 오버라이드(재정의)한다.
        # Child 클래스의 __init__ 메서드를 호출하며, 
        # 이 때 Child 클래스의 cn1과 cn2를 각각 100과 200으로 설정한다.
        super().__init__(100, 200)
        
        self.pn1 = pn1
        self.pn2 = pn2
        
pa = Parent(10, 20)
print(f"pn1 = {pa.pn1}")
print(f"pn2 = {pa.pn2}")
print(f"cn1 = {pa.cn1}")
print(f"cn2 = {pa.cn2}")
````

# 25_다중 상속

### 📎 실습

````python
class Basic_calculator:
    
    def add(self, n1, n2):
        return n1 + n2
    
    def sub(self, n1, n2):
        return n1 - n2
    
    def mul(self, n1, n2):
        return n1 * n2
    
    def div(self, n1, n2):
        return n1 / n2
    
class Developer_calculator:
    
    def mod(self, n1, n2):
        return n1 % n2
    
    def flo(self, n1, n2):
        return n1 // n2
    
    def exp(self, n1, n2):
        return n1 ** n2
    
class Calculator(Basic_calculator, Developer_calculator):
    pass
    
#     상속된 함수만 사용할 것 이기 때문에 n1과 n2는 따로 정의 하지 않는다
#     def __init__(self, n1, n2):  
#         Basic_calculator.__init__(self, n1, n2)
#         Developer_calculator.__init__(self, n1, n2)
        
        
cal = Calculator()
cal.add(20, 4)
````

````python
class Basic_calculatorr:
    
    def __init__ (self, s1, s2):
        print("Basic_calculator")
        
        self.basic_score1 = s1
        self.basic_score2 = s2
        

class Developer_calculatorr:
    
    def __init__(self, s3, s4):
        print("Developer_calculator")
        
        self.dev_score1 = s3
        self.dev_score2 = s4
        
class Calculatorr(Basic_calculatorr, Developer_calculatorr):
    
    def __init__(self, s1, s2, s3, s4):
        print("Calculaotr")
        
        Basic_calculatorr.__init__(self, s1, s2)
        Developer_calculatorr.__init__(self, s3, s4)
        
    def b_add(self):
        result = self.basic_score1 + self.basic_score2
        print(f"basic sum은 {result}")
        
cal = Calculatorr(40, 60, 20, 70)

cal.b_add()
````

# 26_오버라이딩

- 하위 클래스에서 상위 클래스의 메서드를 재정의(override)한다.

### 📎 실습
- 삼각형 넓이를 계산하는 클래스를 만들고 이를 상속하는 클래스에서 getArea()를
- 오버라이딩 해서 출력 결과가 다음과 같을 수 있도록 클래스를 만들어보자.

````python
class Triangle_area:
    
    def __init__(self, w, h):
        
        self.width = w
        self.height = h
        
    def print_tr_info(self):
        print(f"삼각형의 가로길이: {self.width}")
        print(f"삼각형의 세로길이: {self.height}")
        
    def get_area(self):
        return self.width * self.height / 2
    
class Nex_triangle_area(Triangle_area):
    
    def __init__(self, w, h):
        
        super().__init__(w, h)
        
ta = Nex_triangle_area(7, 5)
ta.print_tr_info()
get_area = ta.get_area()

print(f"삼각형의 넓이는: {get_area}cm2")
````

# 27_추상클래스

- 상위 클래스에서 하위 클래스에 메서드 구현을 강요한다.
- 정의되어 있지 않은 class를 추상 클래스라 한다.
- 어떠한 공통적인 기능을 사용하지만 각자 입맛에 맞게 써야 될 때 추상 클래스를 사용한다.

### 📎 실습
- 계산기 추상 클래스를 만들고 이를 이용해서 새로운 계산기 클래스를 만들어 보자.
- 추상 클래스에는 덧셈, 뺄셈, 곱셈, 나눗셈 기능이 선언되어 있어야 한다.

````python
from abc import ABCMeta, abstractmethod

class Calculator(metaclass = ABCMeta):
    
    @abstractmethod
    def add(self, n1, n2):
        pass
    
    @abstractmethod
    def sub(self, n1, n2):
        pass
    
    @abstractmethod
    def mul(self, n1, n2):
        pass
    
    @abstractmethod
    def div(self, n1, n2):
        pass
    
class Develop_cal(Calculator):
    
    def add(self, n1, n2):
        print(n1 + n2)
        
    def sub(self, n1, n2):
        print(n1 - n2)
        
    def mul(self, n1, n2):
        print(n1 * n2)
        
    def div(self, n1, n2):
        print(n1 / n2)
        
    # 필요시 추가적으로 함수 정의.   
    def flo(self, n1, n2):
        print(n1 // n2)
        
cal = Develop_cal()
cal.mul(70, 20)
````

# 29_예외처리

- 예상하지 못한 예외가 프로그램 전체에 영향이 없도록 처리함.

### 📍 try ~ except
- 예외 발생 예상 구문을 try ~ except로 감싼다.

### 📎 실습
- 사용자로부터 숫자 5개를 입력받을 때
- 숫자가 아닌 자료형이 입력되면 예외 처리하는 프로그램을 만들어보자.

````python
nums = []

n = 1

while n < 6:
    try:
        user_n = int(input("input number: "))
    except:
        print("예외발생")
        continue
    nums.append(user_n)
    n += 1
print(f"nums: {nums}")
````

# 30_try ~ except ~ else

### 📍 ~ else
- 예외가 발생하지 않은 경우 실행하는 구문이다.

### 📎 실습
- 사용자로부터 숫자 5개를 입력받아 
- 짝수, 홀수, 실수로 구분해서 각각을 리스트에 저장하는 프로그램을 만들어보자.

````python
even_list = []; odd_list = []; float_list = [];

n = 1

while n < 6:
    
    try:
        user_n = float(input("input number: "))
        
    except:
        print("exception raise!!")
        print("please again")
        
    else:
        if user_n - int(user_n) != 0:
            print("float!!")
            float_list.append(user_n)
            
        else:
            if user_n % 2 == 0:
                print("even!!")
                even_list.append(int(user_n))
                
            else:
                print("odd!!")
                odd_list.append(int(user_n))
        n += 1

print(f"{even_list = }")
print(f"{odd_list = }")
print(f"{float_list = }")
````

# 31_finally

- 예외 발생과 상관없이 항상 실행한다.

### 📎 실습
- 사용자로부터 숫자 5개를 입력받아 
- 짝수, 홀수, 실수와 입력한 모든 데이터를 각각 출력하는 프로그램을 만들어보자.

````python
even_list = []; odd_list = []; float_list = []; data_list = []

n = 1

while n < 6:
    
    try:
        data = input("input number: ")
        f_data = float(data)
        
    except:
        print("exception raise!!")
        print("please again")
        
    else:
        if f_data - int(f_data) != 0:
            print("float!!")
            float_list.append(f_data)
            
        else:
            if f_data % 2 == 0:
                print("even!!")
                even_list.append(int(f_data))
                
            else:
                print("odd!!")
                odd_list.append(int(f_data))
        n += 1
        
    finally:
        data_list.append(data)

print(f"{even_list = }")
print(f"{odd_list = }")
print(f"{float_list = }")
print(f"{data_list = }")
````

### 📎 실습
- 사용자가 문자 메시지를 보낼때 10글자 이하면 SMS로 발송하고, 
- 10글자를 초과하면 MMS로 발송하는 프로그램을 예외처리를 이용해서 만들어보자.

````python
def send_sms(msg):
    
    if len(msg) > 10:
        raise Exception("길이 미달!! SMS 전환 후 발송!!", 1)
    else:
        print("SMS 발송!!")
        
def send_mms(msg):
    
    if len(msg) <= 10:
        raise Exception("길이 초과!! MMS 전환 후 발송!!", 2)
    else:
        print("MMS 발송!!")
        
msg = input("메시지 입력: ")

try:
    send_sms(msg)
    
except Exception as e:
    print(f"e: {e.args[0]}")
    print(f"e: {e.args[1]}")
    
    if e.args[1] == 1:
        send_sms(msg)
        
    elif e.args[1] == 2:
        send_mms(msg)
````

# 33_사용자 Exception 클래스

- Exception 클래스를 상속해서 사용자 예외 클래스를 만들 수 있다.

````python
class Not_user_zero_exception(Exception):
    
    def __init__(self, n):
        super().__init__(f"{n}은 사용할 수 없습니다!!")
        
def div_cal(num1, num2):
    
    if num2 == 0:
        raise Not_user_zero_exception(num2)
    else:
        print(f"{num1} / {num2} = {num1 / num2}")

num1 = int(input("input number1: "))
num2 = int(input("input number2: "))

try:
    div_cal(num1, num2)
except Not_user_zero_exception as e:
    print(e)
````

### 📎 실습
- 관리자 암호를 입력하고 다음 상태에 따라 예외 처리하는 예외 클래스를 만들어보자.





````python
class Password_length_short_exception(Exception):
    pass

class Password_length_long_exception(Exception):
    pass

class Password_wrong_exception(Exception):
    pass

def check_password(password):
    correct_password = "admin1234"
    if len(password) < 5:
        raise Password_length_short_exception(f"{password}: 길이 5미만!!")
        
    elif len(password) > 10:
        raise Password_length_long_exception(f"{password}: 길이 10초과!!")
        
    elif password != correct_password:
        raise Password_wrong_exception(f"{password}: 잘못된 비밀번호!!")
    else:
        print("빙고!")

try:
    password = input("input admin password: ")
    check_password(password)
except Password_length_short_exception as e:
    print(e)
except Password_length_long_exception as e:
    print(e)
except Password_wrong_exception as e:
    print(e)
````

- 파일 모드는 파일을 어떤 목적으로 open 할지 정한다.
>- ‘w’ : 쓰기 전용(파일이 있으면 덮어씌움) 
>- ‘a’ : 쓰기 전용(파일이 있으면 덧붙임)
>-  ‘x’ : 쓰기 전용(파일이 있으면 에러 발생) 
>- ‘r’ : 읽기 전용(파일이 없으면 에러 발생)

# 37_with ~ as문

- with ~ as문을 이용하면 파일 닫기(close)를 생략할 수 있다.

````python
uri = '/Users/seokminlee/Desktop/mose/zero-base/스터디노트/2주차/'

file = open(uri + "test.txt", "w")
file.write("안녕하세요. 이석민입니다.")
file.close()
````

````python
file = open(uri + "test.txt", "r")
str = file.read()
print(str)
````

````python
with open(uri + "test.txt", 'r') as f:
    print(f.read())
````

### 📎 실습
- 로또 번호 생성기 프로그램을 만들고 파일에 번호를 출력해 보자.

````python
import random


uri = '/Users/seokminlee/Desktop/mose/zero-base/스터디노트/2주차/'
def write_number(nums):
    for idx, num in enumerate(nums):
        
        with open(uri + "lotto.txt", 'a') as f:
            if idx < (len(nums) - 2):
                f.write(str(num) + ", ")
                
            elif idx == (len(nums) - 2):
                f.write(str(num))
                
            elif idx == (len(nums) - 1):
                f.write("\n")
                f.write("bonus: " + str(num))
                f.write("\n")
                
r_nums = random.sample(range(1, 46), 7)
print(f"r_nums: {r_nums}")

write_number(r_nums)
````

# 38_writelines()

- writelines( )는 리스트(List) 또는 튜플 데이터를 파일에 쓰기 위한 함수이다.

````python
languages = ["java\n", "c#\n", "python\n", "javascript\n"]

with open(uri + "language.txt", "w") as f:
    f.writelines(languages)
````

````python
with open(uri + "language.txt", "r") as f:
    print(f.read())
````

### 📎 실습
- 딕셔너리에 저장된 과목별 점수를 파일에 저장하는 코드를 작성하자.

````python
score_dic = {"kor": 85, "eng": 90, "mat" : 92, "sci": 79, "his": 82}

uri = '/Users/seokminlee/Desktop/mose/zero-base/스터디노트/2주차/'

for key in score_dic.keys():
    with open(uri + "dic.txt", "a") as f:
        f.writelines(key + " : " + str(score_dic[key]) + "\n")
````

````python
with open(uri + "dic.txt", "r") as f:
    data = f.read()
    print(data)
````

# 39_readlines(), readline()

### 📍 readlines( )
- 파일의 모든 데이터를 읽어서 리스트 형태로 반환한다.

````python
with open(uri + "language.txt", "r") as f:
    data = f.readlines()
    print(data)
````

### 📍 readline( )
- 한 행을 읽어서 문자열로 반환 한다.

````python
with open(uri + "language.txt", "r") as f:
    line = f.readline()
    
    while line != "":
        print(f"line: {line}", end="")
        line = f.readline()
````
