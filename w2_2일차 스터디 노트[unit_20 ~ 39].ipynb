### üìç super(  )
- ÏÉÅÏúÑ ÌÅ¥ÎûòÏä§Ïùò ÏÜçÏÑ±ÏùÑ Ï¥àÍ∏∞ÌôîÌïòÍ∏∞ ÏúÑÌï¥ÏÑú super( )Î•º Ïù¥Ïö©ÌïúÎã§.

````python
class Child:
    
    def __init__(self, cn1, cn2):
        print("print child numbers")
        self.cn1 = cn1
        self.cn2 = cn2


class Parent(Child):
    
    def __init__(self, pn1, pn2):
        print("print parent numbers")
        
        # Child ÌÅ¥ÎûòÏä§Ïùò __init__ Î©îÏÑúÎìúÎ•º Ïò§Î≤ÑÎùºÏù¥Îìú(Ïû¨Ï†ïÏùò)ÌïúÎã§.
        # Child ÌÅ¥ÎûòÏä§Ïùò __init__ Î©îÏÑúÎìúÎ•º Ìò∏Ï∂úÌïòÎ©∞, 
        # Ïù¥ Îïå Child ÌÅ¥ÎûòÏä§Ïùò cn1Í≥º cn2Î•º Í∞ÅÍ∞Å 100Í≥º 200ÏúºÎ°ú ÏÑ§Ï†ïÌïúÎã§.
        super().__init__(100, 200)
        
        self.pn1 = pn1
        self.pn2 = pn2
        
pa = Parent(10, 20)
print(f"pn1 = {pa.pn1}")
print(f"pn2 = {pa.pn2}")
print(f"cn1 = {pa.cn1}")
print(f"cn2 = {pa.cn2}")
````

````python
class Triangle_area:
    
    def __init__(self, w, h):
        
        self.width = w
        self.height = h
        
    def print_tr_info(self):
        print(f"ÏÇºÍ∞ÅÌòïÏùò Í∞ÄÎ°úÍ∏∏Ïù¥: {self.width}")
        print(f"ÏÇºÍ∞ÅÌòïÏùò ÏÑ∏Î°úÍ∏∏Ïù¥: {self.height}")
        
    def get_area(self):
        return self.width * self.height / 2
    
class Nex_triangle_area(Triangle_area):
    
    def __init__(self, w, h):
        
        super().__init__(w, h)
        
ta = Nex_triangle_area(7, 5)
ta.print_tr_info()
get_area = ta.get_area()

print(f"ÏÇºÍ∞ÅÌòïÏùò ÎÑìÏù¥Îäî: {get_area}cm2")
````

# 27_·Ñé·ÖÆ·Ñâ·Ö°·Üº·Ñè·Ö≥·ÜØ·ÑÖ·Ö¢·Ñâ·Ö≥

- ÏÉÅÏúÑ ÌÅ¥ÎûòÏä§ÏóêÏÑú ÌïòÏúÑ ÌÅ¥ÎûòÏä§Ïóê Î©îÏÑúÎìú Íµ¨ÌòÑÏùÑ Í∞ïÏöîÌïúÎã§.
- Ï†ïÏùòÎêòÏñ¥ ÏûàÏßÄ ÏïäÏùÄ classÎ•º Ï∂îÏÉÅ ÌÅ¥ÎûòÏä§Îùº ÌïúÎã§.
- Ïñ¥Îñ†Ìïú Í≥µÌÜµÏ†ÅÏù∏ Í∏∞Îä•ÏùÑ ÏÇ¨Ïö©ÌïòÏßÄÎßå Í∞ÅÏûê ÏûÖÎßõÏóê ÎßûÍ≤å Ïç®Ïïº Îê† Îïå Ï∂îÏÉÅ ÌÅ¥ÎûòÏä§Î•º ÏÇ¨Ïö©ÌïúÎã§.

### üìé Ïã§Ïäµ
- Í≥ÑÏÇ∞Í∏∞ Ï∂îÏÉÅ ÌÅ¥ÎûòÏä§Î•º ÎßåÎì§Í≥† Ïù¥Î•º Ïù¥Ïö©Ìï¥ÏÑú ÏÉàÎ°úÏö¥ Í≥ÑÏÇ∞Í∏∞ ÌÅ¥ÎûòÏä§Î•º ÎßåÎì§Ïñ¥ Î≥¥Ïûê.
- Ï∂îÏÉÅ ÌÅ¥ÎûòÏä§ÏóêÎäî ÎçßÏÖà, Î∫ÑÏÖà, Í≥±ÏÖà, ÎÇòÎàóÏÖà Í∏∞Îä•Ïù¥ ÏÑ†Ïñ∏ÎêòÏñ¥ ÏûàÏñ¥Ïïº ÌïúÎã§.

````python
from abc import ABCMeta, abstractmethod

class Calculator(metaclass = ABCMeta):
    
    @abstractmethod
    def add(self, n1, n2):
        pass
    
    @abstractmethod
    def sub(self, n1, n2):
        pass
    
    @abstractmethod
    def mul(self, n1, n2):
        pass
    
    @abstractmethod
    def div(self, n1, n2):
        pass
    
class Develop_cal(Calculator):
    
    def add(self, n1, n2):
        print(n1 + n2)
        
    def sub(self, n1, n2):
        print(n1 - n2)
        
    def mul(self, n1, n2):
        print(n1 * n2)
        
    def div(self, n1, n2):
        print(n1 / n2)
        
    # ÌïÑÏöîÏãú Ï∂îÍ∞ÄÏ†ÅÏúºÎ°ú Ìï®Ïàò Ï†ïÏùò.   
    def flo(self, n1, n2):
        print(n1 // n2)
        
cal = Develop_cal()
cal.mul(70, 20)
````

# 30_try ~ except ~ else

### üìç ~ else
- ÏòàÏô∏Í∞Ä Î∞úÏÉùÌïòÏßÄ ÏïäÏùÄ Í≤ΩÏö∞ Ïã§ÌñâÌïòÎäî Íµ¨Î¨∏Ïù¥Îã§.

### üìé Ïã§Ïäµ
- ÏÇ¨Ïö©ÏûêÎ°úÎ∂ÄÌÑ∞ Ïà´Ïûê 5Í∞úÎ•º ÏûÖÎ†•Î∞õÏïÑ 
- ÏßùÏàò, ÌôÄÏàò, Ïã§ÏàòÎ°ú Íµ¨Î∂ÑÌï¥ÏÑú Í∞ÅÍ∞ÅÏùÑ Î¶¨Ïä§Ìä∏Ïóê Ï†ÄÏû•ÌïòÎäî ÌîÑÎ°úÍ∑∏Îû®ÏùÑ ÎßåÎì§Ïñ¥Î≥¥Ïûê.

````python
even_list = []; odd_list = []; float_list = [];

n = 1

while n < 6:
    
    try:
        user_n = float(input("input number: "))
        
    except:
        print("exception raise!!")
        print("please again")
        
    else:
        if user_n - int(user_n) != 0:
            print("float!!")
            float_list.append(user_n)
            
        else:
            if user_n % 2 == 0:
                print("even!!")
                even_list.append(int(user_n))
                
            else:
                print("odd!!")
                odd_list.append(int(user_n))
        n += 1

print(f"{even_list = }")
print(f"{odd_list = }")
print(f"{float_list = }")
````

# 33_·Ñâ·Ö°·Ñã·Ö≠·Üº·Ñå·Ö° Exception ·Ñè·Ö≥·ÜØ·ÑÖ·Ö¢·Ñâ·Ö≥

- Exception ÌÅ¥ÎûòÏä§Î•º ÏÉÅÏÜçÌï¥ÏÑú ÏÇ¨Ïö©Ïûê ÏòàÏô∏ ÌÅ¥ÎûòÏä§Î•º ÎßåÎì§ Ïàò ÏûàÎã§.

````python
class Not_user_zero_exception(Exception):
    
    def __init__(self, n):
        super().__init__(f"{n}ÏùÄ ÏÇ¨Ïö©Ìï† Ïàò ÏóÜÏäµÎãàÎã§!!")
        
def div_cal(num1, num2):
    
    if num2 == 0:
        raise Not_user_zero_exception(num2)
    else:
        print(f"{num1} / {num2} = {num1 / num2}")

num1 = int(input("input number1: "))
num2 = int(input("input number2: "))

try:
    div_cal(num1, num2)
except Not_user_zero_exception as e:
    print(e)
````

### üìé Ïã§Ïäµ
- Í¥ÄÎ¶¨Ïûê ÏïîÌò∏Î•º ÏûÖÎ†•ÌïòÍ≥† Îã§Ïùå ÏÉÅÌÉúÏóê Îî∞Îùº ÏòàÏô∏ Ï≤òÎ¶¨ÌïòÎäî ÏòàÏô∏ ÌÅ¥ÎûòÏä§Î•º ÎßåÎì§Ïñ¥Î≥¥Ïûê.





````python
class Password_length_short_exception(Exception):
    pass

class Password_length_long_exception(Exception):
    pass

class Password_wrong_exception(Exception):
    pass

def check_password(password):
    correct_password = "admin1234"
    if len(password) < 5:
        raise Password_length_short_exception(f"{password}: Í∏∏Ïù¥ 5ÎØ∏Îßå!!")
        
    elif len(password) > 10:
        raise Password_length_long_exception(f"{password}: Í∏∏Ïù¥ 10Ï¥àÍ≥º!!")
        
    elif password != correct_password:
        raise Password_wrong_exception(f"{password}: ÏûòÎ™ªÎêú ÎπÑÎ∞ÄÎ≤àÌò∏!!")
    else:
        print("ÎπôÍ≥†!")

try:
    password = input("input admin password: ")
    check_password(password)
except Password_length_short_exception as e:
    print(e)
except Password_length_long_exception as e:
    print(e)
except Password_wrong_exception as e:
    print(e)
````
