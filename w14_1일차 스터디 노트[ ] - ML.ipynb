# 1. PCA

- PCA(주성분 분석, Principal Component Analysis)는 데이터의 차원 축소를 위한 통계적 방법 중 하나입니다.
- 머신러닝, 데이터 분석 및 다양한 응용분야에서 널리 사용되며, 데이터의 주요 패턴을 유지하면서 데이터를 저차원으로 변환하는 데 유용합니다.
### 1️⃣ 변동성
- PCA는 데이터의 최대 변동성을 캡처하려고 합니다. 이것은 데이터의 분산이 가장 큰 방향을 찾는 것을 의미합니다.

### 2️⃣ 주성분
- 변동성이 가장 큰 방향을 "첫 번째 주성분"이라고 부릅니다. 
- 두 번째 주성분은 첫 번째 주성분과 직교하면서 가장 큰 변동성을 갖는 방향입니다. 
- 이러한 방식으로 데이터의 모든 차원에 대해 주성분을 찾을 수 있습니다.

### 3️⃣ 차원 축소
- 원하는 차원 수(예: 2D 또는 3D)로 데이터를 축소하려면 상위 몇 개의 주성분만 선택하면 됩니다.

### 4️⃣ PCA의 주요 단계
1. 데이터를 중앙에 맞춥니다 (평균을 0으로 만듭니다).
2. 데이터의 공분산 행렬을 계산합니다.
3. 공분산 행렬의 고유값과 고유벡터를 찾습니다.
4. 고유값을 내림차순으로 정렬하고, 해당하는 고유벡터를 선택합니다.
5. 상위 k개의 고유벡터를 사용하여 데이터를 k차원으로 변환합니다.
### 5️⃣ PCA의 장점
- 데이터의 차원을 축소하여 시각화가 쉬워집니다.
- 저장 공간 및 계산 시간을 절약할 수 있습니다.
- 과적합을 방지하는 데 도움이 될 수 있습니다.
### 6️⃣ PCA의 단점
- 원래 데이터의 모든 정보를 유지할 수 없습니다.
- 주성분이 항상 데이터에 유용한 특성을 반영한다고는 할 수 없습니다.
