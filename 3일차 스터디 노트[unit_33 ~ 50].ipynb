## 📍 실습

- 다음 반복 실행을 위해서 for문과 while문 중 가장 적합한 구문을 이용해서 프로그램을 만들어 보자.
> - 자동차 바퀴가 한번 구를 때마다 0.15mm씩 마모된다고 한다.
> - 현재의 바퀴 두께가 30mm이고 최소 운행 가능 바쿼 두께가 20mm라고 할 때 앞으로 구를 수 있는 횟수를 구해보자.

```python
wheel_size = 30
minus_wheel = 0.15
rotation_cnt = 0

while wheel_size >= 20: 
    rotation_cnt += 1
    wheel_size -= minus_wheel
    
# 마지막 바퀴를 구른 후 20 미만일 경우 산정. 따라서 1횟수 차감.
print(f"구른 횟수: {rotation_cnt - 1}")
```

# 38. 반복문 제어( continue )

- 반복 실행 중 continue를 만나면 실행을 생략하고, 다음 반복 실행문으로 넘어간다.

```python
for i in range(1, 50):
    if i % 7 != 0:
        continue
        
    print(f"{i}는 7의 배수입니다.")
```

```python
# else의 실행문은 반복문이 종료된 후 실행된다. 궁금!!

cnt = 0
for i in range(1, 100):
    if i % 7 != 0:
        continue
    
    print(f"{i}는 7의 배수이다.")
    cnt += 1

else:
    print(f"99까지의 정수 중 7의 배수는 {cnt}개 이다.")
```

- 1부터 100까지의 정수 중 3과 7의 공배수와 최소 공배수를 출력하자.

```python
min_num = 0

for i in range(1, 100):
    if i % 3 != 0 or i % 7 != 0:
        continue
    
    # 3과 7의 공배수 출력
    print(f"공배수: {i}")
    
    # 3과 7의 공배수 중 최초의 공배수만 반복문 밖의 min_num에 정의된다. 
    if min_num == 0:
        min_num = i

else:
    print(f"최소 공배수: {min_num}")
```

# 39. 반복문 제어( break )

```python
# 반복 실행 중 break를 만나면 반복문을 빠져나온다.

num = 0

while 1:
    print(f"hello")
    
    num += 1
    
    if num >= 5:
        break
```

```python
# 1부터 100까지 정수를 더할 때, 합계가 100이 넘는 최초의 정수 찾기.

sum = 0
first_num = 0
for i in range(1, 100):
    sum += i

    if sum > 100:
        first_num = i
        break
        
print(f"최초의 정수: {first_num}")
    
```

- 10의 팩토리얼(10!)을 계산하는 과정에서 결과값이 50을 넘을 때의 숫자를 구하자.

```python
# result값이 0이면 계속 0이기 때문에, result는 1로 시작한다. 
result = 1
output_num = 0

for i in range(1, 11):
    result *= i
    
    if result > 50:
        output_num = i
        break
        
print(f"넘을 때의 숫자는 {output_num}이다.")
```

- 새끼 강아지 체중이 2.2kg가 넘으면 이유식을 중단하려고 한다. 한번 이유식을 먹을 때 체중이 70g 증가한다고 할 때,
- 예상되는 이유식 날짜를 구하자. (현재 체중은 800g이다.)

```python
max_kg = 22000
plus_kg = 70
current_kg = 800
date = 1

while 1:
    if current_kg >= max_kg:
        break
        
    current_kg += plus_kg
    date += 1
    
print(f"이유식 중단 날짜는 {date} 입니다.")
```

# 49. [연습문제] 연산자(02)

```python
kor = int(input("국어 점수 입력: "))
eng = int(input("영어 점수 입력: "))
mat = int(input("수학 점수 입력: "))
total = kor + eng + mat
avg = round(total / 3, 2)
max_score = max([kor, eng, mat])
min_score = min([kor, eng, mat])

print(f"총점: {total}")
print(f"평균: {avg}")

print("-"*30)
if max_score == kor:
    print(f"최고 점수 과목(점수): 국어({kor})")
elif max_score == eng:
    print(f"최고 점수 과목(점수): 영어({eng})")
elif max_score == mat:
    print(f"최고 점수 과목(점수): 수학({mat})")

if min_score == kor:
    print(f"최저 점수 과목(점수): 국어({kor})")
elif min_score == eng:
    print(f"최저 점수 과목(점수): 영어({eng})")
elif min_score == mat:
    print(f"최저 점수 과목(점수): 수학({mat})")

print(f"최고, 최저 점수 차이: {max_score - min_score}")
print("-"*30)
```

# 50. [연습문제] 연산자(03)

- 시, 분, 초를 입력하면 초로 환산하는 프로그램을 만들어보자.

```python
hou = int(input("시간 입력: "))
min = int(input("분 입력: "))
sec = int(input("초 입력: "))
print(f"{hou * (60 ** 2) + (min * 60) + sec:,}초")
```

- 금액, 이율, 거치기간을 입력하면 복리계산하는 복리계산기 프로그램을 만들어보자.

```python
money = int(input("금액 입력: "))
rate = float(input("이율 입력: "))
term = int(input("기간 입력: "))

target_money = money

for i in range(term):
    target_money += target_money * rate * 0.01

print("-"*30)
print(f"이율: {rate}%")
print(f"원금: {money:,}원")
print(f"{term}년 후 금액: {round(target_money):,}원")
print("-"*30)
```
