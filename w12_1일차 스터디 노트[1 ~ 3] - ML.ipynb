```python
from sklearn.model_selection import train_test_split

X = iris.data[:, 2:]
y = iris.target

# X, y( ) Train dataset  Test dataset .
X_train, X_test, y_train, y_test = train_test_split(X, #   Features, input
                                                    y, #   Labels, output(target)
                                                    test_size=0.2,# (20%) //   test set  (: 0.25)
                                                    stratify=y, #         . Dataset  , (  .)
                                                    shuffle=True, #    () : True - .(), False -  .
                                                    random_state=0 # random seed .
                                                   )
```

```python
X_train.shape
```

###   /    ?

```python
np.unique(y_test, return_counts=True)
```

##   ,   

```python
tree = DecisionTreeClassifier(random_state=0)

tree.fit(X_train, y_train)
```

```python
plot_tree(tree)
```

```python
pred_train = tree.predict(X_train)
```

```python
# 
acc = accuracy_score(y_train, pred_train)
acc
```

```python
from mlxtend.plotting import plot_decision_regions

plt.figure(figsize=(14, 8))
plot_decision_regions(X=X_train, y=y_train, clf=tree, legend=2)
plt.show()
```

```python
# 
pred_test = tree.predict(X_test)
acc = accuracy_score(y_test, pred_test)
acc
```

```python
#   
tree.predict_proba(X_test)
```
